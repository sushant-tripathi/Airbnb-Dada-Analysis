# **Description of the features of Airbnb dataset**

Now our Airbnb dataset has around 49,000 observations in it with 16 columns and it is a mix between categorical and numeric values. Below are the features and their description:

**id and name** - They are simply the listing id and their names.

**host_id and host_name** - They signify the id of host who are associated to the respective listing and their names.

**neighbourhood_group** - It contains the district of the listing.

**neighbourhood** - the neighbourhood of the respective neibourhood groups.

**latitude and longitude** - the geogarphical co-ordinates of respective listings.

**room_types** - listing space type.

**price** - price of listing in US dollars.

**minimum_nights** - amount of nights minimum.

**calculated_host_listings_count** - amount of listing per host.

**availability_365** - number of days when listing is available for booking.

**number_of_reviews** - Total number of reviews.

**reviews_per_month** - average of reviews per month.

**last_review** - date of the last review recorded.

**Note:** Most importantly, the **min price** is showing as **0** and the **max price** as **10,000**. Something isn't right with the data, so need to look into this issue and check for outliers.

```python
# Checking for outliers
df.agg({'price':['mean','median','min','max','count']})
```

# **Observation:** 
The minimum value is set as zero, which does not make any sense because no one will provide free rooms on Airbnb and the maximum value is set as 10,000 which is just seems too high.

Notice how the mean is considerably higher than the median. This is an indication that our data is **skewed**.

Let's use a **boxplot** to get a better understanding of **price distribution**:

```python
# Plotting the boxplot of the price data.
plt.figure(figsize=(15,7))
sns.boxplot(y='price',data=df)
plt.title('Price Distribution',fontweight="bold",fontsize=18)
plt.ylabel('Price')
plt.show()
```

# **Observation:** 
As expected there are outliers present in the data. In order to fix this, we will be using the quantile based flooring and capping.
First, let's print the 10th and 90th percentile of the price column.

```python
# To get the low quantile value
low_quantile = df['price'].quantile(0.10)
low_quantile
```

```python
# To get the high quantile value
high_quantile = df['price'].quantile(0.90)
high_quantile
```

# **Observation:** 
Looks like the 10th percentile is 49USD and the 90th percentile is 269USD. Since we are dealing with Airbnb listings, it is common for a few places to be much more expensive than others, especially luxurious ones.

Now, we can remove the values that don't apply to the specified range – set between 49USD and 269USD because the mean is skewed by outliers.

```python
# Removing the values below 10th percentile and above 90th percentile
df1 = df.drop(df[df['price']<low_quantile].index)
df1 = df.drop(df[df['price']>high_quantile].index)
```

```python
# Lets create a new boxplot and check the result.
plt.figure(figsize=(15,7))
sns.boxplot(y='price',data=df1)
plt.title('Price Distribution',fontweight="bold",fontsize=18)
plt.ylabel('Price')
plt.show()
```
